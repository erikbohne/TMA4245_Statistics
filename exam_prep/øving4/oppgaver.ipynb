```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def S():
    return np.random.normal(loc=241.3, scale=2, size=1)

def T():
    return np.random.normal(loc=7.02, scale=1, size=1)
```

```python
n = 10000

G = np.zeros(n)
for i in range(n):
    G[i] = 2*S() / T()**2

plt.hist(G, bins=100)
```

```python
SD = np.std(G)
MU = np.mean(G)

print("SD = ", SD)
print("MU = ", MU)
```

```python
from scipy.stats import poisson

def X(n, l):
    return (1/n) * sum(poisson.rvs(mu=l, size=n))

r = 100000
l = 12
n = 42
m = 35
alpha = (2*n)/(2*n + m)
beta = (2*m)/(2*m + n)

lambda_hat = np.zeros(r)
for i in range(r):
    lambda_hat[i] = alpha*X(n, l) + beta*X(n, l/2)

plt.hist(lambda_hat, bins=25)
```

```python
import statsmodels.api as sm

sm.qqplot(lambda_hat)
plt.show()
```
