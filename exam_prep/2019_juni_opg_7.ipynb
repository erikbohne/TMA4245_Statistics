# Analyse av Forkastningsregel 1 for Uniformfordeling

Vi starter med Ã¥ definere sannsynlighetstetthetsfunksjonen for en uniform fordeling og visualisere dens form for forskjellige verdier av \( \theta \).

## Visualisering av Uniform Densitet


```python
import numpy as np
import matplotlib.pyplot as plt

def uniform_density(x, theta):
    if theta <= x <= theta + 1:
        return 1
    else:
        return 0

theta_values = [0, 0.5, 1, 1.5]
x_values = np.linspace(-1, 3, 400)

plt.figure(figsize=(12, 6))
for theta in theta_values:
    plt.plot(x_values, [uniform_density(x, theta) for x in x_values], label=f'Theta = {theta}')

plt.title('PDF of Uniform Distributions for Different Theta Values')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

## Beregning og Visualisering av Type I-feil og Testens Styrke

FÃ¸rst beregner vi sannsynligheten for Type I-feil for Forkastningsregel 1, som forkaster \( H_0 \) hvis \( X_1 > 0.95 \).


```python
# Sannsynlighet for Type I-feil
P_type_I_error = 1 - 0.95

# Definering av theta-verdier for Ã¥ beregne styrken av testen
theta_range = np.linspace(0.01, 1, 100)

# Beregning av styrken av testen for Forkastningsregel 1
power_of_test = [max(0, 1 - (0.95 - tau)) if 0.95 >= tau else 1 for tau in theta_range]

plt.figure(figsize=(12, 6))
plt.plot(theta_range, power_of_test, label='Test Styrke som Funksjon av Theta (Ï)', color='blue')
plt.title('Test Styrke som Funksjon av Theta (Ï) for Forkastningsregel 1')
plt.xlabel('Theta (Ï)')
plt.ylabel('Sannsynlighet for Ã¥ Forkaste H0')
plt.grid(True)
plt.show()
```
