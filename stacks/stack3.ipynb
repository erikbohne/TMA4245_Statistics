```python
import numpy as np
u = np.random.uniform()
def y(u):
    return 2*np.sqrt(u)

print(" u = 0.15 -> y =", y(0.15))
```

```python
import matplotlib.pyplot as plt

def F(y):
    return (y/2)**2

plt.figure()
plt.title("CDF")
plt.xlabel("y")
plt.ylabel("F(y)")
plt.plot(np.linspace(0,2,100), F(np.linspace(0,2,100)))
plt.show()
```

```python
print("P(Y <= 0.6) =", F(0.6))
```
